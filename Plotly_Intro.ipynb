```python
# Standard plotly imports
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
# Using plotly + cufflinks in offline mode
import cufflinks
cufflinks.go_offline(connected=True)
init_notebook_mode(connected=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('Bank_Personal_Loan_Modelling-1.xlsx',sheetname='Data')
```

```python
df.pivot(columns='Personal Loan', values='Mortgage').iplot(
kind='box',
yTitle='Value of House mortgage($000)',
title='Personal Loan-Mortgage')
```

```python
df.pivot(columns='Personal Loan', values='Income').iplot(
kind='box',
yTitle='Income in($000)',
title='Personal Loan-Income')
```

```python
import plotly.figure_factory as ff
```

```python
corrs = df.corr()
figure = ff.create_annotated_heatmap(
z=corrs.values,
x=list(corrs.columns),
y=list(corrs.index),
annotation_text=corrs.round(2).values,
showscale=True)
```

```python
df['Income'].iplot(kind='hist', xTitle='Income',
yTitle='count', title='Income Distribution')
```

```python
df[['Income', 'Mortgage']].iplot(
kind='hist',
histnorm='percent',
barmode='overlay',
title='Income-Mortgage')
```

```python
figure = ff.create_scatterplotmatrix(
df[['Income','CCAvg','Mortgage']],
diag='histogram',
index='Income')
```
